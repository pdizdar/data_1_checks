```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# selected column headers we want to use
new_headers = ['SCHOOL YEAR', 'COUNTY NAME', 'DISTRICT NAME', 'SCHOOL NAME', 'Total Spending per Student - Federal Funds', 'Total Spending per Student - State/Local Funds',
       'Total Spending per Student - All Fund Sources']

# import spending_per_student_2021 data into dataframes
df = pd.read_csv("spending_per_student_2021.csv", header=0)

# remove unwanted columns keep selected headers
dataset = pd.DataFrame(df, columns=new_headers)

#if county name is equal to Kenton and print
kenton_county_schools = dataset.loc[dataset['COUNTY NAME'] == 'KENTON']

#from kenton, if district = kenton county school district and print
kenton_county_school_district = kenton_county_schools.loc[kenton_county_schools['DISTRICT NAME'] == 'Kenton County']


#plotting data
a = kenton_county_school_district['SCHOOL NAME']
b = kenton_county_school_district['Total Spending per Student - Federal Funds']
c = kenton_county_school_district['Total Spending per Student - State/Local Funds']
d = kenton_county_school_district['Total Spending per Student - All Fund Sources']

#creating 
plt.subplot(131)
plt.plot(a, b, color='red')
plt.title('Federal Spending per Student')
plt.xlabel('School Name')
plt.ylabel('Total Spending per Student - Federal Funds')

plt.subplot(132)
plt.bar(a, c, color='orange')
plt.title('State/Local Spending Per Student')
plt.xlabel('School Name')
plt.ylabel('Total Spending per Student - State/Local Funds')

plt.subplot(133)
plt.scatter(a, d, color='green')
plt.title('All Fund Sources per Student')
plt.xlabel('School Name')
plt.ylabel('Total Spending per Student - All Fund Sources')

'''#creating subplots
fig, axs = plt.subplots(3)
axs[0].plot(a, b, color='red')
axs[0].set_title('Federal Spending per Student')
axs[0].set_xlabel('School Name')
axs[0].set_ylabel('Total Spending per Student - Federal Funds')

axs[1].plot(a, c, color='orange')
axs[1].set_title('State/Local Spending Per Student')
axs[1].set_xlabel('School Name')
axs[1].set_ylabel('Total Spending per Student - State/Local Funds')

axs[2].plot(a, d)
axs[2].set_title('All Fund Sources per Student')
axs[2].set_xlabel('School Name')
axs[2].set_ylabel('Total Spending per Student - All Fund Sources')'''
#idx = df.columns
county = dataset.groupby(['COUNTY NAME', 'DISTRICT NAME'])[['Total Spending per Student - All Fund Sources', 'Total Spending per Student - Federal Funds', 'Total Spending per Student - State/Local Funds']].sum()
'''fed_spend = df.groupby(['COUNTY NAME','DISTRICT NAME'])[['Total Spending per Student - State/Local Funds','Total Spending per Student - Federal Funds']].sum()
'''
#print(fed_spend)
#print(county)
#print(idx)
#print(dataset)
#print(kenton_county_schools)
print(kenton_county_school_district)

```
