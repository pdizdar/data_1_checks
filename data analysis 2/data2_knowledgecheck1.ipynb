Import pandas and requests library to pull in data from Makeup API.

```python
import pandas as pd
import requests
```

```python
# get makeup api and print the message
makeup = requests.get('http://makeup-api.herokuapp.com/api/v1/products.json')
if makeup:
    print('Request is successful.')
else:
    print('Request returned an error.')
```

```python
# print the data in json format
makeup.json()[0]
```

```python
# to see the type of data
print(type(makeup.json()))
```

Using pandas built in function 'json_normalize' to flatten our data.

```python
# normalize and  flatten the json data into a dataframe
makeup_api = pd.json_normalize(makeup.json())
makeup_api.head(4) # print the first 4 rows of the dataframe
```

```python
makeup_api.tail() # print tail of dataframe
```

```python
# info about the columns of data 
makeup_api.info() 
```

By looking at the datatype above, datatype for price is 'object' which clearly needs to be float. This shows the shape of the data where we have 931 rows and 19 columns. Also shows that some of the columns have null values.

```python
# changing datatype of price from object to float
makeup_api.price = makeup_api.price.astype(float)
makeup_api.price.dtype
```

```python
# get summary statistics of the data
makeup_api.describe()
```

```python
mean = makeup_api['rating'].mean().round(2)
min_ratings = makeup_api['rating'].min()
max_ratings = makeup_api['rating'].max()

print(f'The average ratings for all the makeup brand is {mean}')
print(f'The minimum ratings for all the makeup brand is {min_ratings}')
print(f'The maximum ratings for all the makeup brand is {max_ratings}')
```

Get all the rows where the price is greater than and equal to 30 to see the highend makeup products.

```python
highend_makeup = makeup_api[makeup_api['price'] >= 30]
highend_makeup

```

```python
# info on lowerend makeup 
lowerend_makeup = makeup_api[makeup_api['price'] <= 10]
lowerend_makeup
```
