```python
import pandas as pd
```

```python
data = pd.read_excel('StudentParticipationAggregate_EndofYear.xlsx')
#data.head()
```

```python
#check the shape of data
data.shape
# has 700 rows and 100 columns
```

```python
#column names to see which i want to keep
data.columns
```

```python
# creating new dataframe with only the columns i need
new_headers = ['School Year', 'District Number', 'districtName',
       'Learning Model for Date Range', 'Total Students',
       'Total Days Scheduled', 'Participation Days', 'Non Participation Days', 'Report Date Range']
dataset = pd.DataFrame(data, columns=new_headers)
dataset
```

```python
dataset.shape
# 700 rows and 9 columns
```

```python
# Renaming DistrictName column
dataset.rename(columns={'DistrictName': 'District Name'}, inplace=True)
dataset
```

```python
# by looking at data we can see that bottom two rows have nan values so we will get rid of it
# drop any rows that has nan values
dataset.dropna(how='any', inplace=True)
dataset.shape
# 2 rows with nan has been dropped
```
